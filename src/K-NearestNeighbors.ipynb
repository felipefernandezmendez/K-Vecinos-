```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
from sklearn.neighbors import NearestNeighbors
```

```python
data = pd.read_csv('/workspaces/K-NearestNeighbors/data/processed/clean-films.csv')
```

```python
data.head(3)
```

```python
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(data["description"])   # Vectoriza los datos de texto
```

```python
model = NearestNeighbors(n_neighbors=5, metric='cosine')
model.fit(tfidf)
```

```python
# Función para recomendar películas similares

def recommend_similar_movies(movie):
    movie_index = data[data['title'] == movie].index[0]
    _, indices = model.kneighbors(tfidf[movie_index])    # Encuentra los índices de las películas más similares

    # Imprime los títulos de las películas más similares
    print("Película base:", data['title'].iloc[movie_index])
    print("Películas recomendadas:")
    for i, idx in enumerate(indices[0]):
        print(f"{i+1}. {data['title'].iloc[idx]}")
```

```python
data['title']
```

```python
input_movie = "Pirates of the Caribbean: At World's End"
recommendations = recommend_similar_movies(input_movie)
```

```python
movie = "Avatar"
recommend_similar_movies(movie)
```

Optimizar un modelo k-NN implica ajustar los hiperparámetros para mejorar su rendimiento. Pasos que a seguir para optimizar un modelo k-NN:

Seleccionar el valor de k: El hiperparámetro principal en k-NN es el número de vecinos, k. Puedes probar diferentes valores de k y evaluar cómo afecta el rendimiento del modelo en los datos de prueba utilizando técnicas de validación cruzada o dividir los datos en conjunto de entrenamiento y prueba.

Normalización de características: Dado que k-NN se basa en la distancia entre los puntos de datos, la normalización de las características puede mejorar el rendimiento del modelo. Asegúrate de escalar o normalizar tus características antes de ajustar el modelo.

Distancia ponderada: En lugar de asignar el mismo peso a todos los vecinos más cercanos, puedes ponderar la distancia según su importancia relativa. Esto puede mejorar el rendimiento del modelo, especialmente cuando hay desequilibrios en la distribución de clases.

Selección de métrica de distancia: Experimenta con diferentes métricas de distancia, como la distancia euclidiana, la distancia de Manhattan o la distancia de Mahalanobis, para encontrar la que se ajuste mejor a tus datos.

Búsqueda de rejilla (Grid Search): Utiliza la búsqueda de rejilla para explorar de manera sistemática diferentes combinaciones de hiperparámetros y encontrar la configuración óptima que maximice la métrica de evaluación, como la precisión o el área bajo la curva ROC.

Para evaluar si el modelo k-NN está bien implementado, puedes seguir estos pasos:

Validación cruzada: Utiliza técnicas de validación cruzada para evaluar el rendimiento del modelo en múltiples particiones de los datos. Esto te dará una estimación más robusta del rendimiento del modelo en datos no vistos.

Métricas de evaluación: Calcula métricas de evaluación como precisión, recall, F1-score y matriz de confusión para comprender cómo se comporta el modelo en diferentes clases y en conjunto de datos de prueba.

Curva ROC y área bajo la curva (AUC): Si estás trabajando con un problema de clasificación binaria, puedes trazar la curva ROC y calcular el área bajo la curva (AUC) para evaluar el rendimiento del modelo en diferentes umbrales de clasificación.

Comparación con otros modelos: Compara el rendimiento del modelo k-NN con otros algoritmos de aprendizaje automático en el mismo conjunto de datos para determinar si es competitivo en términos de precisión y eficiencia.
