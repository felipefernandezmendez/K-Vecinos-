```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
from sklearn.neighbors import NearestNeighbors
```

```python
data = pd.read_csv('/workspaces/K-NearestNeighbors/data/processed/clean-films.csv')
```

```python
data.head(3)
```

```python
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(data["description"])   # Vectoriza los datos de texto
```

```python
model = NearestNeighbors(n_neighbors=5, metric='cosine')
model.fit(tfidf)
```

```python
# Función para recomendar películas similares

def recommend_similar_movies(movie):
    movie_index = data[data['title'] == movie].index[0]
    _, indices = model.kneighbors(tfidf[movie_index])    # Encuentra los índices de las películas más similares

    # Imprime los títulos de las películas más similares
    print("Película base:", data['title'].iloc[movie_index])
    print("Películas recomendadas:")
    for i, idx in enumerate(indices[0]):
        print(f"{i+1}. {data['title'].iloc[idx]}")
```

```python
input_movie = "Pirates of the Caribbean: At World's End"
recommendations = recommend_similar_movies(input_movie)
```
